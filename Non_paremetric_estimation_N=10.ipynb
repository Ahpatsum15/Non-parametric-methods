```python

#number of realizations 
R= np.arange(1,20)
#ER graph 
G=nx.erdos_renyi_graph(10,0.3)
S=nx.adjacency_matrix(G) #we define the shift 
S
```

```python
np.shape(S)
```

```python
#On definie le filtre H
h=np.random.random_sample((4,)) #we define shift's coefficients 
H = np.zeros(10)
S_0 = np.identity(10)
S_2= S@S
S_3=S_2@S
H= h[0]*S_0+ h[1]*S+h[2]*S_2+h[3]*S_3
```

```python
H
```

```python
H_her= H.getH()
C_x= H@H_her
```

```python
C_x
```

```python
p,V=np.linalg.eig(C_x)
V_her=V.getH()
```

```python
p #PSD  
V_her
```

```python
V@V.getH() #verification that V is unitary 
```

```python
V_her=np.array(V_her)
V_her
```

```python
def module_vec(x): 
  n=np.size(x)
  y=np.zeros((n,1))
  for i in range(n) :
    y[i][0]=x[i]**2
  return y
```

```python
w1=np.random.normal(0,1,10)
module_vec(w1)
```

```python
MSE= []
for i in R: 
  p_pg =np.zeros((10,1))
  for j in range(1,i+1): 
    w=np.random.normal(0,1,10)
    w=np.transpose(w)
    y=H@w
    y=np.transpose(y)
    p_pg= p_pg + module_vec(np.dot(V_her,y))
  b_pg = p_pg/i - p 
  MSE.append(sp.linalg.norm(b_pg) + 2*sp.linalg.norm(p)**2/i)
MSE = np.array(MSE)  
```

```python
#the normalize MSE
norm_mse = MSE/np.linalg.norm(MSE)
norm_mse
```

```python
#code pour plot 
plt.plot(R,norm_mse, color ='b')
#On compare la valeur du MSE théorique avec celle qui a été calculé par des moyens epiriques. 
plt.xlabel("number of realizations")
plt.ylabel('normalised MSE' )
plt.legend(['normalized MSE'])
plt.title('ER graph with N=10 and p=0.3')
plt.show()
```
